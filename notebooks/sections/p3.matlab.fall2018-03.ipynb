### Section 3.1: Image Processing Basics.

We can load an image into MATLAB by supplying the path to the `imread` function.  Save `rgb.png` to the variable `color_circles`.  **Make sure to put the semicolon at the end of the following command.**

```matlab
color_circles = imread('../../data/images/examples/rgb.png');
```

We can now view the image using `imshow`:

```matlab
imshow(color_circles)
```

#### Discussion

* What are the dimensions of `color_circles`?  What do they represent?
* What information is saved in `color_circles(:, :, 1)`? 
* Type `color_circles` into the console.  What is being printed out?  (*Press `Ctrl` + `C` to force-stop MATLAB*)


* `rgb.png` is a 1000 by 1000 `png` picture.  However, when we look at the size of the image in MATLAB, we get

```matlab
format compact
size(color_circles)
```

The third dimension of length 3 represent the images color channels. 

* The information stored in `color_circles(:, :, 1)` is the images red channel.  Similarly, `color_circles(:, :, 2)` is its green channel and `color_circles(:, :, 3)` is its blue channel.
* Each pixel of the image is represented by a tuple `(r, g, b)`, where `r` is that pixels red value, `g` its green value, and `b` its blue value.  When we type `color_circles` into the console, each of those pixel values gets put out to the console.  

(Read more about the RGB color model [here](https://en.wikipedia.org/wiki/RGB_color_model))

We will now extract the red values of `color_circles`.  Save the red channel to a variable named `red_channel` and display it using `imshow`:

```matlab
red_channel = color_circles(:, :, 1);
imshow(red_channel)
```

What was the expected output?  What happened?

We can get our desired output by adding in empty green and blue channels, then layering `red_channel` with the buffers.  We want our buffer to be a matrix of zeros with the same dimensions as our original picture.  We will use the `zeros` and `size` commands to achieve this:

```matlab
buffer_channel = zeros(size(color_circles, 1), size(color_circles, 2));
imshow(buffer_channel)
```

We will now use the `cat` command to concatenate our three layers along the dimension corresponding to the color value:

```matlab
just_red = cat(3, red_channel, buffer_channel, buffer_channel);
imshow(just_red)
```
